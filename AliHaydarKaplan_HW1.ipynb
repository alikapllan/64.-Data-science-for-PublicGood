```python
#1
"""
Write a code that modifies the given matrix M (with 3 rows and 4 columns), by swapping the (whole) columns 0 and 1.

M = [[11, 12, 13, 14], [21, 22, 23, 24], [31, 32, 33, 34]]

The matrix M above should become the matrix below;

M = [[12, 11, 13, 14], [22, 21, 23, 24], [32, 31, 33, 34]]

"""

M = [[11, 12, 13, 14], [21, 22, 23, 24], [31, 32, 33, 34]]
print(M)

for i in range(len(M)):
    M[i][0],M[i][1] = M[i][1],M[i][0]
    
print(M)
```

```python
#2 
"""
Find the maximal element of the matrix M and print the row number and column number of the maximal element as a tuple.

M = [[0, 3, 2, 4], [2, 7, 5, 9], [6, 1, 2, 3]]

The output of the code should be row number and column number of the maximal element as a tuple. (1,3)
"""
M = [[0, 3, 2, 4], [2, 7, 5, 9], [6, 1, 2, 3]]


import numpy as np
M = [[0, 3, 2, 4], [2, 7, 5, 9], [6, 1, 2, 3]]
arr = np.array(M)
result = np.where(arr==np.amax(arr))
tuple_version = tuple(zip(result[0],result[1]))
print(tuple_version)
```

```python
#4
"""
We have a dictionary containing the math grades and the student number of 5 students. 
Calculate the mean of the grades and then prints out the name of the students who are above and below that average(and if any print that are equals to the mean as well):

For example if the grades are

Jane: 30

Pat: 40

Denis: 50

Max: 60

Lara: 70

Then the mean is 50

your program should print:

*"Grade(s) of Jane and Pat are below the average"*

*"Grade(s) of Denis is equal to the mean"*

*"Grade(s) of Max and Lara are above the average"*

"""
students = {"Jane":{"grade":90,"student_number":4012}, 
          "Pat":{"grade":81,"student_number":4052},
          "Denis":{"grade":70,"student_number":4089},
          "Max":{"grade":72,"student_number":4322},
          "Lara":{"grade":74,"student_number":4042}}

gradesList = []
for student in students :
    gradesList.append(students[student]["grade"])

average = sum(gradesList) / len(gradesList)


below = []
equal = []
above = []

for student in students :
    if students[student]["grade"] > average : 
        above.append(student)
    elif students[student]["grade"] < average : 
        below.append(student)
    else: 
        equal.append(student)


print("Grade(s) of ", ','.join(above) if above else 'no one ', " is/are above the average." )
print("Grade(s) of ",','.join(below) if below else 'no one ', " is/are below the average." )
print("Grade(s) of ",','.join(equal) if equal else 'no one ', " is/are equal the average." )

#value_1 if condition else value_2

```
